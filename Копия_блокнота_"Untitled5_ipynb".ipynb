1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю
стоимость за один кв. метр общей площади

1.2. Создать новый признак delta_renovated, который будет содержать разницу в
годах между<br> годом реновацией дома и годом постройки дома.<br> Если реновации
дома не было, то в новом признаке поставьте 0.

```python
import numpy as np
```

```python
df['delta_renovated'] = np.where(df['yr_renovated'] == 0, 0, df['yr_renovated']-df['yr_built'])
```

```python
df.head()
```

1.3 Создайте признаки года продажи, месяца продажи

```python
df['yr_sold'] = df['date'][0:3]
```

```python
df.head()
```

```python
def partline(my_str:str):
  my_str = my_str[0:4]
  return my_str
```

```python
df['yr_sold'] = df['date'].apply(partline)
```

```python
df.head()
```

```python
def partline46(my_str:str):
  my_str = my_str[4:6]
  return my_str
```

```python
df['month_sold'] = df['date'].apply(partline46)
```
