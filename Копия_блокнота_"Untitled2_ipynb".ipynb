```python
def gb_count(my_str:str):
  my_str = my_str.split()[0]
  # print(my_str)
  if my_str.endswith("GB"):
    return int(my_str.replace("GB",""))
  elif my_str.endswith("TB"):
    return 1024*int(float(my_str.replace("TB","")))
```

```python
gb_count("1.0TB HDD")
```

```python
df['Memory_Amount'] = df["Memory"].apply(gb_count)
```
