```python
dfh.waterfront.nunique()
```

```python
sea_view = dfh.waterfront.value_counts()
print("Количество домов без вида на набережную (0) и с видом на набережную(1)")
print(sea_view)
```

```python
plt.figure(figsize=(10, 8))
expl = [0.2, 0.7]
plt.pie(sea_view.values,
        autopct = '%1.1f%%',
        labels = sea_view.index,
        explode = expl)
plt.title('С видом на набережную')
plt.legend(sea_view.index, loc = 'best');
```

```python
dfh.floors.nunique()
```

```python
dfh.condition.nunique()
```

```python
print("Распределение домов по их состоянию:")
condition = dfh.condition.value_counts()
print(condition)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x = condition.index,
            y = condition.values,
            palette='tab20b')
plt.xticks(rotation=0)
plt.title('Рапределение домов по их состоянию');
```

Выводы:
* Подавляющее большинство домов без вида на набережную.
* Этажность большего количества домов лежит в пределах  1 - 2 этажа.
* Состояние большинства домов среднее. Далее их число снижается <br> примерно в 2.5 раза с увеличением качества состояния на 1 пункт.

####Условие 3: 3 задача####
Исследуйте, какие характеристики недвижимости влияют на

стоимость недвижимости, с применением не менее 5 диаграмм из урока. <br>Анализ сделайте в формате storytelling: дополнить каждый график письменными<br>выводами и наблюдениями.

```python
plt.figure(figsize =(8, 6))
sns.lineplot(x = dfh['sqft_living'], y = dfh['price'])
plt.title('Корреляция между ценой и размером жилой площади')
plt.xlabel('размер жилой площади')
plt.ylabel('цена')
```

Стоимость недвижимости линейно растет с ростом жилой площади, с ростом размера жилой площади и ее цены общее  количество недвижимости падает.

```python
plt.figure(figsize =(8, 6))
sns.boxplot(y = dfh['price'], x = dfh['floors'])
plt.title('Зависимость цены от этажности')
plt.xlabel('этажность')
plt.ylabel('цена');
```

Самая высокая цена у домов с этажностью 2.5.  У них самая большая медиана. Дома с этажностью 3.5 имеют самую низкую  стоимость. В диапазоне от 1.0 до 2.5 наблюдается рост цены в зависимости от этажности, далее наблюдается обратная тенденция.

```python
plt.figure(figsize =(8, 6))
sns.boxplot(y = dfh['price'], x = dfh['waterfront'])
plt.title('Зависимость цены от расположения на набережной')
plt.xlabel('располагается (1) или нет (0) на набережной')
plt.ylabel('цена');
```

Дома, расположенные на набережной имеют более высокую цену, в этом сегменте недвижимости цены более стабильные (меньшее  количество выбросов). Цены на дома не раполоенные на набережной более низкие, но наблюдается больший разброс в ценах (больше выбросов).

```python
plt.figure(figsize =(8, 6))
sns.boxplot(y = dfh['price'], x = dfh['view'])
plt.title('Зависимость цены от вида')
plt.xlabel('качество вида (баллы)')
plt.ylabel('цена');
```

Цена недвижимости возрастает с ростом качества вида.

```python
plt.figure(figsize =(8, 6))
sns.lineplot(x = dfh['yr_built'], y = dfh['price'])
plt.title('Зависимость цены от года постройки')
plt.xlabel('цена')
plt.ylabel('год постройки');
```

Цены на дома, построенные  до 1930 года самые высокие и могут сравниться только с ценами на новостройки (2020 год). Цены на недвижимость с 1930 до 1980 года низкие - цены на недвижимость в новостройках выше, чем на вторичном рынке.

Возможные причины:

1. Старые дома строились по индивидуальным проектам и с более интересными и качественными архитектурными решениями.
2. Новые дома - скорее всего жилье первичного рынка, амортизация еше не велика, на могие элементы кострукции домов еще распространяется гарантия застройщика.
3  Дома средней возрастной категории - вторичный рынок, частично утерян товарный вид, возможно нуждаются или скоро будут нуждаться в ремонте.
